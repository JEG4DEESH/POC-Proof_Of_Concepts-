**Handling Missing values**

```python
df.columns
```

```python
df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]=df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']].replace(0,np.NaN)
```

**In this dataset the data collection team has put some values as a zero instead of leave it as a NaN values. so what we are doing is Handling those values as per the domain importance.**

```python
# It returns the total sum no of null values on Feature wise in numbers
# before handling NaN values

df.isnull().sum()
```

**These are all the missing values percentage for the independent feature. If the percentage goes below 3 percentage we can drop those feature as per the domain importance. Also if the percentage goes above 30 percentage we can drop those values**

```python
print("Percentage of missing values in Glucose is : ",5/768*100)
print('Percentage of missing values in BloodPressure is : ',35/768*100)
print("Percentage of missing values in SkinThickness is : ",227/768*100)
print("Percentage of missing values in Insulin is : ",374/768*100)
print('Percentage of missing values in BMI is : ',11/768*100)
```

```python
#Filling those null values using mean max

col1=['SkinThickness','Insulin']
df[col1]=df[col1].fillna(df.mean())
```

```python
col2=['Glucose']
df[col2]=df[col2].fillna(df.mean())
```

```python
col3=['BloodPressure','BMI']
df[col3]=df[col3].fillna(df.max())
```

X & Y SPLIT

```python
x=df.iloc[:,:-1]
```

```python
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=38)
```

```python
print('xtrain',xtrain.shape)
print('xtest',xtest.shape)
print('ytrain',ytrain.shape)
print('ytest',ytest.shape)
```

**MODEL BUILDING : Support Vector Machine**

```python
# Import the Support Vector Classifier class
from sklearn.svm import SVC

# Create a Support Vector Classifier object
svc = SVC()

# Train the classifier on some data
svc.fit(xtrain, ytrain)

# Predict the labels of some new data
y_predict=svc.predict(xtest)
```

**Evaluation Matrices**

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(ytest,y_predict)
```

```python
cm
```

```python
score=svc.score(xtest,ytest)
```

```python
score
```

```python
accuracy_score(ytest,y_predict)
```

```python
print(classification_report(ytest,y_predict))
```

**MODEL BUILDING : DecisionTreeClassifier**

```python
# Import the DecisionTreeClassifier class
from sklearn.tree import DecisionTreeClassifier

# Create a decision tree classifier object
clf = DecisionTreeClassifier()

# Train the classifier on some data
clf.fit(xtrain, ytrain)

# Predict the labels of some new data
y_dc_predict = clf.predict(xtest)
```
